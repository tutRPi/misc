# Test API: Send samples to local running server
Run server locally and then `npx localtunnel --port 8000` - use URL below

```python
import requests
import json

host = 'https://every-owls-rush-77-190-28-150.loca.lt'
url = host + "/receipt"


class PdEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, pd.Timestamp):
            return str(obj)
        return json.JSONEncoder.default(self, obj)


# index = 3584979 
# row = df.iloc[index]
# if True:
for index, row in df.sample(n=5).iterrows():
  obj = row.to_dict()
  json_str = json.dumps(obj, cls=PdEncoder)

  print("Sending row with id =", index, "json =", json_str)
  x = requests.post(url, data=json_str)
  
  if x.status_code != 201:
    print("Response:", x.content)
  else:
    print("SUCCESS")
  print()
```
